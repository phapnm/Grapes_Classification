```python
INPUT_SHAPE = (224, 224 ,3)
N_CLASSES = 4
```

```python
class MultiTask:
    @staticmethod
    def build(inputShape = INPUT_SHAPE, classes = N_CLASSES, finalAct = 'softmax'):
        # DepthWiseCONV => CONV => RELU => POOL
        model = Sequential()
        
        
        model.add(Conv2D(32, (3, 3), padding='same', activation='relu'))
        model.add(BatchNormalization(axis=-1))
        model.add(Conv2D(32, (3, 3), padding='same', activation='relu'))
        model.add(BatchNormalization(axis=-1))
        model.add(MaxPooling2D(pool_size=(3,3)))
        model.add(Dropout(0.25))
        
        # (CONV => RELU) * 2 => POOL
        model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
        model.add(BatchNormalization(axis=-1))
        model.add(Conv2D(64, (3, 3), padding='same', activation='relu'))
        model.add(BatchNormalization(axis=-1))
        model.add(MaxPooling2D(pool_size=(2,2)))
        
        # (CONV => RELU) * 4 => POOL
        model.add(Conv2D(128, (3, 3), padding='same', activation='relu'))
        model.add(BatchNormalization(axis=-1))
        model.add(Conv2D(128, (3, 3), padding='same', activation='relu'))
        model.add(BatchNormalization(axis=-1))
        model.add(Conv2D(128, (3, 3), padding='same', activation='relu'))
        model.add(BatchNormalization(axis=-1))
        model.add(Conv2D(128, (3, 3), padding='same', activation='relu'))
        model.add(BatchNormalization(axis=-1))
        model.add(MaxPooling2D(pool_size=(2, 2)))
        
        # firt (and only) set of FC => RELU layers
        model.add(Flatten())
        model.add(Dense(1024, activation='relu'))
        model.add(BatchNormalization())
        model.add(Dropout(0.4))
        
        # softmax classifier
        model.add(Dense(classes))
        model.add(Activation(finalAct))
        return model
```

```python
model = MultiTask.build(inputShape = INPUT_SHAPE, classes = N_CLASSES, finalAct = "sigmoid")
```

```python
EPOCHS = 75
INIT_LR = 0.001
BATCH_SIZE = 32
opt = Adam(lr=INIT_LR, decay=INIT_LR / EPOCHS)
```

```python
#imagePaths = (sorted(_list_images(root_dir='./dataset')))
imagePaths = sorted(list(paths.list_images("datasets")))
data = []
labels = []
```

```python
# Đọc data đầu vào từ imagePaths và gán label cho data 
for imagePath in imagePaths:
    image = cv2.imread(imagePath)
    #Resize ảnh 224X224
    image = cv2.resize(image, (224, 224))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = img_to_array(image)
    data.append(image)
    #gán dữ liệu labels
    label = imagePath.split(os.path.sep)[-2].split("_")
    labels.append(label)
```

```python
# scale [0, 1]
data = np.array(data, dtype="float")/255.0
```

```python
# Chuyển label của ảnh thành ma trận nhị phân
mlb = MultiLabelBinarizer()
y = mlb.fit_transform(labels)
```

```python
# Lữu trữ mlb.pkl file
f = open('mlb.pkl', "wb")
f.write(pickle.dumps(mlb))
f.close()
```

```python
mlb.transform([("NinhThuan", "Red")])
```

```python
#(X_train, X_val, y_train, y_val)= MultilabelStratifiedShuffleSplit()
#(X_train, y_train, X_val, y_val) = iterative_train_test_split(data, y, test_size = 0.2)
(X_train, X_test, y_train, y_test) = train_test_split(data, y, test_size=0.4, random_state=42,stratify=y,shuffle=True)
(X_test, X_val, y_test, y_val) = train_test_split(X_test, y_test, test_size=0.5, random_state=42,stratify=y_test,shuffle=True)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
print(X_val.shape, y_val.shape)
```

```python
aug = ImageDataGenerator(rotation_range=25, width_shift_range=0.1,
                        height_shift_range=0.1, shear_range=0.2, zoom_range=0.2,
                        horizontal_flip=True, fill_mode="nearest")
```

```python
model = MultiTask.build(inputShape=INPUT_SHAPE, classes=N_CLASSES, finalAct = 'sigmoid')
#inputShape = INPUT_SHAPE, classes = N_CLASSES, finalAct = 'softmax'
```

```python
# Hàm tạo metric F1-score
def recall_m(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    recall = true_positives / (possible_positives + K.epsilon())
    return recall

def precision_m(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    return precision

def f1_m(y_true, y_pred):
    precision = precision_m(y_true, y_pred)
    recall = recall_m(y_true, y_pred)
    return 2*((precision*recall)/(precision+recall+K.epsilon()))
```

```python
model.compile(loss="binary_crossentropy", optimizer=opt, metrics=["accuracy",f1_m])
```

```python
history = model.fit_generator(
    aug.flow(X_train, y_train, batch_size=BATCH_SIZE),
    validation_data=(X_val, y_val),
    steps_per_epoch=len(X_train) // BATCH_SIZE,
    epochs=EPOCHS, verbose=1)
```

```python
model.save("model_grapes_multitask_learning.h5")
```

```python
#Predict bộ test và chuyển sang dạng ma trận nhị phân như y_test
y_pred=model.predict(X_test,batch_size=10,verbose=1)
y_pred = (y_pred > 0.5) 
np.argmax(y_pred, axis=1)
y_pred.astype(int)
```

```python
# Chuẩn hóa đầu vào để predict khi kiểm tra ảnh ngẫu nhiên trên mạng
import requests

def _downloadImage(url):
    re = requests.get(url)
    img = np.asarray(bytearray(re.content), dtype="uint8")
    img = cv2.imdecode(img, cv2.IMREAD_COLOR)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```

```python
def _predict_image(image, model, mlb):
    (w,h,c) = image.shape
    # Nếu resize with = 400 thì height resize sẽ là
    height_rz = int(h*400/w)
    # Resize lại ảnh để hiển thị
    output = cv2.resize(image, (height_rz, 400))
    # Resize lại ảnh để dự báo
    image = cv2.resize(image, IMAGE_DIMS[:2])/255.0
    # Dự báo xác suất của ảnh
    prob = model.predict(np.expand_dims(image, axis=0))[0]
    # Trích ra 2 xác suất cao nhất
    argmax = np.argsort(prob)[::-1][:2]
    # Show classes và probalility ra ảnh hiển thị
    for (i, j) in enumerate(argmax):
        # popup nhãn và xác suất dự báo lên ảnh hiển thị
        label = "{}: {:.2f}%". format(mlb.classes_[j], prob[j] * 100)
        cv2.putText(output, label, (10, (i*30) + 25),
                   cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,255,0),2)
   # for (label, p) in zip(mlb.classes_, prob):
    #    print("{}: {:.2f}%".format(label, prob*100))
    # Hiển thị ảnh dự báo
    plt.figure(figsize=(8,16))
    plt.axis('Off')
    plt.imshow(output)        
```
